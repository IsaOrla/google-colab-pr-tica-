```python
import numpy as np
import pandas as pd
```

```python
df_dados_paises = pd.read_csv('Atividade cap 11 dados_1997_2011_paises_csv.csv', sep =';', encoding = 'latin1', decimal = ',')
df_dados_paises.head(30)
```

```python
df_dados_paises.groupby('pais').agg(Contagem = pd.NamedAgg('pais','count')).reset_index()
```

```python
df_dados_paises.dtypes
```

Métrica (Mínima, Máxima, Média)

```python
#essas barras deixam o código identado, junção
df_dados_paises.groupby('pais') \
                                .agg(min_idh = pd.NamedAgg('idh', 'min'),
                                   max_idh = pd.NamedAgg('idh', 'max'),
                                   media_idh = pd.NamedAgg('idh', 'mean')) \
                                   .reset_index()
```

Moda

```python
df_dados_paises.groupby('pais')['idh'].apply(lambda x: x.mode().iloc[0]).to_frame().reset_index()
```

Mediana

```python
from pandas.core.groupby.generic import NamedAgg
df_dados_paises.groupby('pais').agg(mediana = pd.NamedAgg('idh', 'median')).reset_index()
```

Percentis

```python
df_dados_paises.groupby('pais')['idh'].apply(lambda x: x.quantile([0.05,0.25,0.5,0.75,0.95])).to_frame().reset_index().rename(columns={'level_1':'percentil'})
```

```python
from pandas.core.groupby.generic import SeriesGroupBy
percentis = df_dados_paises.groupby('pais')['idh'] \
                                  .apply(lambda x: x.quantile([0.05, 0.95])) \
                                  .to_frame() \
                                  .reset_index() \
                                  .rename(columns={'level_1' : 'percentil'})
```

```python
percentis.pivot(index='pais', columns='percentil', values='idh') \
                           .reset_index() \
                           .rename(columns={0.05 : 'p_5', 0.95 : 'p_95'})
```

####Medidas de variabilidade

Desvio médio absoluto

```python
df_dados_paises.groupby('pais') \
                      .agg(desvio_absoluto = pd.NamedAgg('idh', 'mad')) \
                      .reset_index()
```

Variancia e Desvio Padrao

```python
df_dados_paises.groupby('pais') \
              .agg(variancia = pd.NamedAgg('idh', 'var'),
                   desvio_padrao = pd.NamedAgg('idh', 'std')) \
                   .reset_index()
```
