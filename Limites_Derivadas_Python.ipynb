###Estudo de Limites da função logística (Regressão Logística)

```python
# analise limites
x = smp.Symbol('x')
funcao = 1 / (1+smp.exp(-x))
# limite tendendo a zero
smp.limit(funcao, x, 0)
# limite tendendo ao infinito
smp.limit(funcao, x, smp.oo)
# limite tendendo ao infinito negativo
smp.limit(funcao, x, -smp.oo)
# grafico
smp.plot(funcao)
```

###Estudo de Limites da função linear (Regressão Linear)

```python
# analise limite
x = smp.Symbol('x')
beta0 = smp.Symbol('beta0')
beta1 = smp.Symbol('beta1')
```

```python
funcao = beta0 + beta1*x
# limite tendendo a zero
smp.limit(funcao, x, 0)
```

```python
funcao = beta0 + beta1*x
# limite tendendo ao infinito
smp.limit(funcao, x, smp.oo)
```

###Derivadas

Derivada com potenciação

```python
x = smp.Symbol('x')
x
```

```python
f = x**2
f
```

```python
smp.diff(f)
```

```python
# graficos
funcao_original = x**2
funcao_derivada = smp.diff(funcao_original)
funcao_original, funcao_derivada
```

```python
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
```

```python
# grafico funcao original
y = [funcao_original(i) for i in range(-100, 100)]
plt.plot(y)
```

```python
# grafico funcao derivada
y_d = [funcao_derivada(i) for i in range(-100, 100)]
plt.plot(y_d)
```

Derivada com multiplicação

```python
# obter a derivada
x = smp.Symbol('x')
f = (x**2 + 3) * (2*x + 1)
f
```

```python
f_d = smp.diff(f)
smp.simplify(f_d)
```

```python
# graficos
funcao_original = (x**2 + 3) * (2*x + 1)
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
```

```python
# funcao original
y = [funcao_original(i) for i in range(-100, 100)]
plt.plot(y)
```

```python
# funcao derivada
y_d = [funcao_derivada(i) for i in range(-100, 100)]
plt.plot(y_d)
```

Derivadas em divisão



```python
# obter a derivada
x = smp.Symbol('x')
f = (3*x**2 + 1) / (2*x - 1)
f_d = smp.diff(f)
smp.simplify(f_d)
# graficos
funcao_original = (3*x**2 + 1) / (2*x - 1)
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
# grafico funcao original
y = [funcao_original(i) for i in range(-10, 10)]
plt.plot(y)
# grafico funcao derivada
y_d = [funcao_derivada(i) for i in range(-10, 10)]
plt.plot(y_d)
```

Derivadas em logaritmo

```python
# obter derivada
x = smp.Symbol('x')
f = x * smp.log(x)
f_d = smp.diff(f)
smp.simplify(f_d)
# graficos
funcao_original = x * smp.log(x)
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
# grafico funcao original
y = [funcao_original(i) for i in range(1, 10)]
plt.plot(y)
# grafico funcao derivada
y_d = [funcao_derivada(i) for i in range(1, 10)]
plt.plot(y_d)
```

Derivadas em exponencial

```python
# obter a derivada
x = smp.Symbol('x')
f = x * smp.exp(2*x)
f_d = smp.diff(f)
smp.simplify(f_d)
# graficos
funcao_original = x * smp.exp(2*x)
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
# grafico funcao original
y = [funcao_original(i) for i in range(0, 10)]
plt.plot(y)
# grafico funcao derivada
y_d = [funcao_derivada(i) for i in range(1, 10)]
plt.plot(y_d)
```
