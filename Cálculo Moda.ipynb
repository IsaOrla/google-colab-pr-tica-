```python
import pandas as pd
```

```python
dados_2019 = pd.read_csv('vendas_linha_petshop_2019.csv', sep =';', encoding = 'latin1', decimal = ',')
```

```python
import statistics

from statistics import mode

# Dados de exemplo para valor_total_bruto e regiao_pais
valor_total_bruto = dados_2019['valor_total_bruto']
regiao_pais = dados_2019['regiao_pais']

# Filtrando os dados apenas para a região Norte
dados_regiao_norte = [valor for valor, regiao in zip(valor_total_bruto, regiao_pais) if regiao == 'Norte']

# Calculando a moda
moda_regiao_norte = statistics.mode(dados_regiao_norte)

print(f"A moda para a região Norte é: {moda_regiao_norte}")
```

```python
# Calculando o percentual de dados faltantes para cada coluna
percent_missing_quantidade = (dados_2019['quantidade'].isnull().sum() / len(dados_2019)) * 100
percent_missing_valor_total_bruto = (dados_2019['valor_total_bruto'].isnull().sum() / len(dados_2019)) * 100

# Imprimindo os resultados
print(f"Percentual de dados faltantes para 'quantidade': {percent_missing_quantidade:.2f}%")
print(f"Percentual de dados faltantes para 'valor_total_bruto': {percent_missing_valor_total_bruto:.2f}%")
```
